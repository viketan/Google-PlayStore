```python
#segregate categorical and neumerical features

num_features=[features for features in df.columns if df[features].dtypes != 'O']
```

```python
num_features
```

```python
categorical_features=[features for features in df.columns if df[features].dtypes =='O']
```

```python
categorical_features
```

```python
df['App'].value_counts(normalize=True)*100
```

```python
#proportion of count in the categorical columns

for col in categorical_features:
    #print(col)
    print(df[col].value_counts())
    print("**********************")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
```

```python
num_features
```

```python
import seaborn as sns
sns.kdeplot(x=df['Rating'])
```

```python
plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(num_features)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[num_features[i]],shade=True, color='r')
    plt.xlabel(num_features[i])
    plt.tight_layout()
```

```python
# insight we can find out over here that rating and year are the left skewd rest of the variable are right skewd
```

```python
# categorical columns
plt.figure(figsize=(20, 15))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)
category = [ 'Type', 'Content Rating']
for i in range(0, len(category)):
    plt.subplot(2, 2, i+1)
    sns.countplot(x=df[category[i]],palette="Set2")
    plt.xlabel(category[i])
    plt.xticks(rotation=45)
    plt.tight_layout() 
```
